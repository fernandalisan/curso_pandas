# Análise exploratória dos dados:

```python
#Valor médio de aluguel por tipo de imóvel e ordenado pela coluna valor:
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
#Plotando média de aluguel por tipo de imóvel:
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
#kind> tipo de gráfico/ figsize> tamanho da figura> color> cor 
df_preco_tipo.plot(kind= 'barh', figsize= (14,10), color = 'purple');
```

#### Ao analisar o gráfico ficou claro que existem muitos tipos de imóveis e muitos deles são para fins comerciais. Como não é o foco da nossa imobiliária os alugueis de imóveis deste tipo vamos excluir todos os de tipo comencial.

## Removendo os imóveis comerciais

```python
dados.Tipo.unique()
```

```python
#Retirando os imóveis comerciais do df: 
imoveis_comerciais = ['Conjunto Comercial/Sala','Prédio Inteiro','Loja/Salão',
                      'Galpão/Depósito/Armazém', 'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial','Chácara', 'Loteamento/Condomínio',
                      'Pousada/Chalé','Hotel', 'Indústria', 'Sítio']

df = dados.query('@imoveis_comerciais not in Tipo')
df.Tipo.unique()
```

```python
#Plotando média de aluguel por tipo de imóvel residencial:
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind= 'barh', figsize= (14,10), color = 'purple');
```

```python
#Percentual de imóveis no df:
#Contando quantas vezes os tipos de imóvel aparecem na base de dados

#contagem dos valores únicos da coluna 'Tipo'
df.Tipo.value_counts()
#normalize=True >> em percentual
df.Tipo.value_counts(normalize=True)
# .to_frame()>>transformar a series em df
df.Tipo.value_counts(normalize=True).to_frame()
#.sort_values('Tipo')>> ordenar do menor para o maior da coluna 'Tipo'
df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

```python
#Plotando média de aluguel por tipo de imóvel residencial:
df_percentual_tipo = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_percentual_tipo.plot(kind= 'bar', figsize= (14,10), 
                        color = 'blue', xlabel= 'Tipos', 
                        ylabel= 'Percentual');
```

#### De acordo com o gráfico vemos que a qt de apartamentos  é muito maior que todos os outros tipos de imóveis. Sendo assim vamos excluir os outros tipos de imóveis para que possamos analisar somento os apartamentos.

## Selecionando apenas os apartamentos

```python
df = df.query('Tipo == "Apartamento"')
df.head()
```

# Tratando dados nulos:

```python
#total de valores nulos por coluna
df.isnull().sum()
```

```python
#substituindo valores nulos por zero
df = df.fillna(0)
df.isnull().sum()
```

# Removendo registros inconsistentes (valores iguais a 0)

```python
#selecionando linhas com valor igual a zero 
registros_a_remover = df.query('Valor == 0 | Condominio == 0').index
df.drop(registros_a_remover, axis=0, inplace=True)
```

```python
#Removendo a coluna 'Tipo' do data frame
df.drop('Tipo', axis=1, inplace=True)
df.head()
```

# Filtros 

```python
#selecionando apartamentos que possuem 1 quarto e aluguel menor que 1200 reais. 
df_1 = df.query('Quartos == 1 and Valor < 1200')
df_1
```

```python
#Selecionando apartamentos que possuem pelo menos 2 quartos, aluguel menor que 3000 e area maior que 70 m2
df_2 = df.query('Quartos >= 2 and Valor < 3000 and Area > 70')
df_2
```

# Criando colunas numéricas

#### Criando coluna valor_por_mes. Esta coluna deve conter gastos mensais de cada imóvel, incluindo aluguel e condomínio. 


```python
url = ('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv')

#Lendo o arquivo csv, passando o parâmetro 'sep = ;'para separar os dados por;
pd.read_csv(url, sep = ';')
dados = pd.read_csv(url, sep = ';')
dados
```

```python
#Criando a coluna valor_por_mes>>  coluna Valor + coluna Condomínio
dados['Valor_por_mes'] = dados.Valor + dados.Condominio 
dados.head()
```

```python
#Criando a coluna valor_por_ano >> coluna Valor_por_mes * 12 + coluna IPTU
dados['Valor_por_ano'] = (dados.Valor_por_mes * 12) + dados.IPTU
dados.head()
```

```python
#Criando coluna Categórica descrição (concateando):
dados['Descrição'] = dados.Tipo + ' em ' + dados.Bairro + ' com ' + \
                    dados.Quartos.astype(str) + ' quartos(s) ' + ' e '+ \
                    dados.Vagas.astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
#criando coluna suites:
dados['Possui_suites'] = dados.Suites.apply(lambda x: 'Sim'if x>0 else 'Não')
dados.head()
```

```python
##salvando o arquivo 
dados.to_csv('dados_completos_dev.csv', index=False)
pd.read_csv('dados_completos_dev.csv')
dados.head()
```
